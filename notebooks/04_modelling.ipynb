Training four regression models on the cleaned GKL housing dataset:

| Model | Role | Tuning |
|-------|------|--------|
| **Ridge** | Linear baseline | Fixed alpha |
| **LightGBM** | Gradient boosting (speed) | Optuna, 50 trials, 5-fold CV |
| **XGBoost** | Gradient boosting (balanced) | Optuna, 50 trials, 5-fold CV |
| **CatBoost** | Gradient boosting (robust) | Optuna, 50 trials, 5-fold CV |

```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))
```

```python
from src.modelling import run_pipeline

results = run_pipeline(n_trials=50)
```

```python
import pandas as pd

summary = pd.DataFrame([
    {"Model": r["Model"], "R²": r["R2"], "MAE": r["MAE"], "RMSE": r["RMSE"]}
    for r in results
]).set_index("Model")
summary.style.format("{:.4f}").highlight_max(subset=["R²"]).highlight_min(subset=["MAE", "RMSE"])
```
