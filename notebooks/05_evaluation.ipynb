# 05 - Model Evaluation

Visual diagnostics for the trained regression models. All figures are pre-generated by the
modelling pipeline and loaded from `outputs/figures/`.

```python
from pathlib import Path
from IPython.display import Image, display, Markdown

fig_dir = Path.cwd().parent / 'outputs' / 'figures'
```

## Model Comparison

Side-by-side R², MAE, and RMSE across all four models. All metrics are computed on
log-transformed prices, so differences in the fourth decimal place are meaningful.

```python
Image(filename=str(fig_dir / '08_model_comparison.png'), width=900)
```

## Actual vs Predicted

Points hugging the diagonal indicate accurate predictions. Deviation at the extremes
(very cheap or very expensive properties) is expected — tree-based models cannot
extrapolate beyond the training range.

```python
Image(filename=str(fig_dir / '09_actual_vs_predicted.png'), width=900)
```

## Residual Plots

Residuals should be randomly scattered around zero with no visible pattern.
Systematic curvature would suggest missing non-linear terms or interaction effects.

```python
Image(filename=str(fig_dir / '10_residual_plots.png'), width=900)
```

## Error Stratification by Price Decile

RMSE broken down by price decile (D1 = cheapest 10%, D10 = most expensive 10%).
The U-shaped pattern — higher errors at both extremes — is characteristic of
tree-based models on skewed targets.

```python
Image(filename=str(fig_dir / '11_error_stratification.png'), width=900)
```

## Feature Importance

Split-based importance from each tree ensemble. Built-up area dominates, followed
by structural features (bathrooms, rooms) and proximity to key infrastructure.

```python
Image(filename=str(fig_dir / '12_feature_importance.png'), width=900)
```

## SHAP Summary (Best Model)

SHAP values show each feature's contribution to individual predictions.
Colour represents feature value (red = high, blue = low). Features are ranked
by mean absolute SHAP value across the test set.

```python
Image(filename=str(fig_dir / '13_shap_summary.png'), width=900)
```
