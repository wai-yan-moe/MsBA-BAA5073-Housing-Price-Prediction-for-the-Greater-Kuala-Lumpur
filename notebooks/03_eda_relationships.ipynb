## Key Takeaways

1. **Built-up area is the strongest driver** -- clear positive relationship with price
2. **Location matters** -- properties closer to KLCC, malls, and transit command higher prices
3. **Property type segments the market** -- bungalows sit at a premium; flats cluster at the lower end
4. **Multicollinearity is present** but tree-based ensemble models are robust to it
5. **Non-linearity is evident** -- linear models will underfit this data
